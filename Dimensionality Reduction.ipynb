# Multidimensional Scaling (MDS) 
    차원을 줄이기 위해 여러 데이터들을 주축에 직교하게 projection한다. 줄여진 차원에서의 데이터 간의 거리를 루트((a-b)**2)라고하면 원래 차원에서의 거리와 차이가 가장 적게해야 올바르게 차원 축소를 한 것이다(차원 축소를 하되, 데이터 간의 거리는 최대한 동일하게 유지)
    그러기 위해서는 sum((원래 차원에서의 데이터 간 거리 - 줄인 차원에서의 데이터 간 거리)**2)을 최소화 해야하고 이것은 gradient descent로 풀 수 있다.

    원래 차원의 각 데이터의 좌표가 주어져서 거리를 구할 수 있을 때는 위의 방식처럼 한다. 이를 Metric MDS 라고 한다.
    그에 반해 각 데이터의 좌표가 주어지지 않고, 데이터 간 거리의 대소만 주어졌을 때는, 즉 데이터 간 거리의 순서만 주어졌을 때 -> 줄여진 차원에서의 데이터 간 거리의 순서도 원래 차원을 따라야 하는데 이때를 NonMetric MDS 라고 한다.
 

# Sammon Mapping
    MDS에서는 기본적으로 데이터간의 거리가 먼 것에 대해 더 중점적으로 최적화를 수행한다. 예를 들어 원래 차원에서의 a,b데이터와 축소 차원에서의 a,b데이터의 거리를 각각 1000, 996이라고하고 원래차원에서의 c,d데이터와 축소 차원에서의 c,d데이터의 거리를 각각 2, 3이라고 한다면, sum(각각 차원에서의 데이터 간 거리의 차이의 제곱)은 거리가 먼 데이터에서 원래 차원과 축소 차원의 차이가 클 확률이 더 높기 때문에 거리가 먼 데이터를 중점적으로 최적화를 하기 마련이다.
    그러나 상대적으로 보면 1000과 996의 차이 4는 1000을 기준으로는 오류가 0.004이지만 2와 3의 차이 1은 2를 기준으로 0.5이므로 거리가 큰 데이터에 대해 중점적으로 최적화를 수행하는 MDS를 변형해야한다. 즉 거리가 가까운 데이터를 낮은 차원으로 옮겼을 때도 그 거리를 잘 보존하도록 MDS를 변형해야함.
    거리가 짧은 데이터를 보존하는 것에 중점을 두는 방식이 Sammon Mapping임. 각 거리의 차이의 제곱에 원래 차원에서의 거리로 나눠져서 sum을 함.



# Cosine Similarity
        벡터 간의 유사도를 거리(Uclidean distance)로 봤을 때는 document의 길이에 영향을 많이 받는다. 벡터 간 유사도는 그 벡터가 각 성분 중 상대적으로 어떤 성분이 더 우세한지가 중요하고 벡터간 유사도를 보는데 있어서 document의 길이는 크게 중요치 않다. document의 길이가 길어서 벡터의 길이가 길어지면 성분의 비율이 비슷하여 유사도가 높게 나와야하는데 벡터간 거리가 멀어지므로 유사도가 낮게 나올 수 있다는 말이다. 따라서 벡터 간 각도로 유사도를 판단하면 벡터의 길이와 관계없이 어떤 성분이 벡터를 이루었는지를 중심으로 유사도를 파악할 수 있다. 
        벡터 간의 유사도를 거리가 아닌 각도로 보면, 벡터간의 각도가 작을수록 유사도가 높다고 볼 수 있다. 데이터 a,b의 각도가 90도 정도고 a,c의 각도가 10도라고 하면 a,b와의 유사도 보다 a,c간 유사도가 더 높다. 세타를 각도라고 했을 때, 세타ab > 세타ac -> cos세타ab < cos세타 ac 가 된다. (cos은 각도가 작아질수록 가장 큰 값인 1에 수렴함). 따라서 cos 세타 값이 클수록 유사도가 높다라고 판단. 이를 Cosine Similarity 라고 한다.

        벡터 간 내적 = 벡터 간 길이 * cos 세타
        cos 세타 = 벡터 간 내적 / 벡터 간 길이
        위의 식으로 MDS 식을 distance가 아닌 cosine similarity 기반으로 변경할 수 있다. 거리의 부작용으로 만들었던 Sammon Mapping에서의 가중치인 1/원래차원의 데이터 거리도 cosine similarity를 사용하면 쓰지 않아도 된다.
        그러면 식은 sum((cos원래 차원의 벡터의 각도 세타 - cos축소된 차원의 벡터의 각도 세타)**2) 이렇게 구성할 수 있다.
