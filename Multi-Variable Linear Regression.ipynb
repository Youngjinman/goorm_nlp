# Multi-Variable Rregression
    y = 세타0 + 세타1x1 + 세타2x2 + 세타3x3 + ... + 세타nxn 일 때(세타는 가중치, xn feature별 데이터)
    [1,x1,x2,...,xn]벡터와 [세타0,세타1,세타2,...세타n]벡터의 내적으로 나타낼 수 있음

## Feature Scaling
    feature마다 단위가 다름 -> ex) feature가 방의 크기(x1), 방의 개수(x2) 라고 하면 x1는 1000단위, x2는 1단위 임. 각각의 변수에 붙는 가중치를 생각하면 x1에 붙는 세타1은 조금만 변해도 전체 크기가 큰 폭으로 변하지만 x2에 붙는 세타2가 조금만 변하면 그 변화 폭은 상대적으로 미미. 기하학적으로 보면 세타1 축을 기준으로 loss function을 자르면 기울기가 급하게 내려가는 모양이지만, 세타2 축을 기준으로 자르면 완만하게 내려가는 모양 -> 세타1으로 편미분하면 절대값이 크게 나올 것이고 learning rate과 곱해서 gradient descent를 하면 크게 움직일 것임. 세타2로 편미분하면 0에 가깝게 나올 것이고 learning rate과 곱해서 gradient descent를 하면 조금씩 움직일 것임.-> 이렇게 되면 최저점까지 가는데 세타1에 대해서는 급하고 세타2에 대해서는 느리니 지그재그로 계속 움직이며 최저점으로 향할 것 -> 최저점으로 일직선으로 쭈욱 향하도록 하는 것이 더 계산적 이점이 있음
    또한 실습에서와 같이 feature scaling을 하지 않으면 gradient descent를 할 때 loss 즉, cost function의 값이 무한대로 발산할 수 있다.
    
    Feature Scaling -> feature의 단위를 normalize하여 세타1, 세타2, ... , 세타n 어느 축으로 봐도 경사가 비슷한 loss function을 구할 수 있음
    feature 단위를 normalize(feature에 대한 모든 데이터를 -1~1 사이로 설정)하려면 feature별 데이터의 최대값과 최소값을 1, -1로 설정해야함. 함수 ax+b를 각각의 데이터가 통과했을 때 -1, 1이 나와야하므로 x자리에 데이터를 각각 넣어서 연립방정식을 풀면 a,b나오고 함수를 구할 수 있음. 그 함수로 모든 데이터를 -1~1 사이로 설정할 수 있음. -> 최소값과 최대값으로 normalize하면 outlier에 취약하게 됨.
    위의 방법보다 많이 쓰는 normalize 방법 -> 정규분포로 맞추는 것 -> 각데이터에 평균값을 빼고 표준편차로 나눠주면 평균=0, 분산=1인 표준정규분포 생성. 모든 데이터들이 그 표준 정규분포상에 위치.
